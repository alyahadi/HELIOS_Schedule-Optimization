### Split multiple tasks

```python
df['Brief Description of Technical Intervention'].unique().tolist()
```

```python
import re
import pandas as pd

# List of phrases that contain separators (&, +, /) but should NOT be split
PROTECTED_PHRASES = [
    "Schedule inspection 1200FH & 3 years Inspection",
    "Finding engine beyond GTA capability & will be send for Level 4 or overhaul repair",
    "Swap of engine box between 5033 and 5034",
    "Transfer box and preservation",
    "Pre Induction report MO2, SERIAL NUMBER: 20514",
    "Pre Induction report MO3, SERIAL NUMBER: 26491",
    "MO3 SERIAL NUMBER 8425 Calender Restoration.",
    "MO3 SERIAL NUMBER 8426 Calender Restoration.",
    "Pre Induction report MAKILA 2A1, ESN13003",

]   

def split_tasks(text):
    if not isinstance(text, str):
        return []

    # Standardize S/E and S/N first
    text = re.sub(r'\bS/E\b', 'STANDARD EXCHANGE', text)
    text = re.sub(r'\bS/N\b', 'SERIAL NUMBER', text)

    # Remove leading engine references (e.g. "ARRIEL 2C ESN: 24433,")
    # These are not tasks â€” just engine identifiers prepended to the actual task
    text = re.sub(r'^ARRIEL\s+\w+\s+ESN[:\s]*\d+\s*,\s*', '', text, flags=re.IGNORECASE)

    # Protect "Do Not Split" phrases
    # We replace spaces/symbols in protected phrases with a safe placeholder
    for phrase in PROTECTED_PHRASES:
        if phrase in text:
            # Create a placeholder version (spaces to underscores, remove split chars later)
            placeholder = phrase.replace(" ", "___SPACE___").replace("&", "___AND___").replace(",", "___COMMA___")
            text = text.replace(phrase, placeholder)

    # Check for numbered list pattern
    # Pattern looks for: Start of line OR space, followed by Digit(s), a Dot, and a Space
    number_pattern = r'(?:^|\s)\d+\.\s+'
    
    tasks = []
    
    # If we find "1. " or " 2. " etc., we use ONLY this to split
    if re.search(number_pattern, text):
        raw_tasks = re.split(number_pattern, text)
        tasks = [t.strip() for t in raw_tasks if t.strip()]
        
    # No numbered pattern found
    else:
        # Proceed with normal splitting rules
        # Convert symbols to common delimiter (comma)
        text = text.replace('&', ', ').replace(' + ', ', ').replace('/', ', ')
        text = re.sub(r'\s+AND\s+', ', ', text, flags=re.IGNORECASE)
        text = re.sub(r'\s+and\s+', ', ', text, flags=re.IGNORECASE)
        
        # Split by comma
        tasks = [t.strip() for t in text.split(',') if t.strip()]
    
    # Restore Protected Phrases
    final_tasks = []
    for t in tasks:
        t = t.replace("___SPACE___", " ").replace("___AND___", "&").replace("___COMMA___", ",")
        final_tasks.append(t)
        
    return final_tasks

# Apply splitting
print("Splitting tasks...")
task_lists = df['Brief Description of Technical Intervention'].apply(split_tasks)

# Find max number of tasks needed
max_tasks = task_lists.apply(len).max()
print(f"Maximum number of tasks found: {max_tasks}")

# Create Task columns
for i in range(max_tasks):
    df[f'Task_{i+1}'] = task_lists.apply(lambda x: x[i] if i < len(x) else None)

print("Splitting complete. Task columns created.")
df.head()

# Display specific examples to verify
cols_to_show = ['Brief Description of Technical Intervention'] + [f'Task_{i+1}' for i in range(min(4, max_tasks))]
print(df[cols_to_show].tail(50))

output_filename = 'technical_intervention_split.xlsx'
df[cols_to_show].to_excel(output_filename, index=False)
print(f"Success! Data saved to {output_filename}")
```

```python
import re
import pandas as pd

def standardize_task1(row):
    """
    Standardizes Task_1 based on its own text AND the original description.
    """
        
    task_text = str(row['Task_1'])
    original_desc = str(row['Brief Description of Technical Intervention'])
    
    # Convert to upper case and strip whitespace
    text = task_text.upper().strip()
    orig_upper = original_desc.upper()
    
    # Basic typo corrections
    text = re.sub(r'\bREPLACEMNET\b', 'REPLACEMENT', text)
    text = re.sub(r'\bREPLACEMENET\b', 'REPLACEMENT', text)
    text = re.sub(r'\bREPLACED\b', 'REPLACEMENT', text)
    text = re.sub(r'\bSTANDART\b', 'STANDARD', text)
    text = re.sub(r'\bASSESMENT\b', 'ASSESSMENT', text)
    text = re.sub(r'\bCHK\b', 'CHECK', text)
    text = re.sub(r'\bINSP\b', 'INSPECTION', text)
    text = re.sub(r'\bBOROSCOPE\b', 'BORESCOPE', text)
    text = re.sub(r'\bF\.C\.U\b', 'FCU', text)
    
    # Remove "CARRIED OUT", "SUPPORT", and "THE" in specific contexts
    text = re.sub(r'\bCARRIED\s+OUT\b', '', text)
    text = re.sub(r'\bSUPPORT\b', '', text)
    text = text.replace("TEST OF THE ", "TEST OF ")

    # TU 81 Rule
    text = re.sub(r'APPLICATION\s+TU\s*81\s+(?:SPARE\s+)?DECU\s+SN\s*\d+', 'TU 81', text)
    
    # Handle Receiving Inspection variations
    text = re.sub(r'\bENGINE\s+RECEIVING\s+INSPECTION\b', 'ENGINE INSPECTION', text)
    text = re.sub(r'\bRECEIVING\s+INSPECTION\b', 'ENGINE INSPECTION', text)
    
    # Keyword mapping
    if text == "RECEIVING":
        text = "ENGINE INSPECTION"
        
    if text == "KANBAN":
        text = "KANBAN REPORT"
        
    if text == "SUPERVISION":
        text = "SCHEDULE INSPECTION"
        
    if text == "EXTERNAL":
        text = "EXTERNAL INSPECTION"

    # TU / TF STANDARDIZATION
    text = re.sub(r'\b(TU|TF)\s*(\d+)(?:\s+APPLICATION)?\b', r'\1 \2', text)

    # MODULE STANDARDIZATION
    text = re.sub(r'\bMODULE\s+0?(\d)\b', r'MO\1', text) 
    text = re.sub(r'\bMOD\s+0?(\d)\b', r'MO\1', text)
    text = re.sub(r'\bM0(\d)\b', r'MO\1', text)
    text = re.sub(r'\bMODULE\s+(MO\d+)\b', r'\1', text)

    # REPLACEMENT / INSTALLATION PATTERNS
    text = re.sub(r'^REPLACEMENT\s+(MO\d+)$', r'\1 REPLACEMENT', text)

    text = text.replace("INSTALLATION OF THE ", "INSTALLATION OF ")
    text = text.replace("INSTALLATION HMU", "HMU INSTALLATION ")

    flip_pattern = r'\b(CERTIFICATION|INSTALLATION|INSPECTION|REPLACEMENT)\s+(?:OF|ON)\s+(.+)'
    match = re.search(flip_pattern, text)
    if match:
        action = match.group(1)
        obj = match.group(2)
        text = f"{obj} {action}"

    # SPECIFIC TEXT REPLACEMENTS
    if "PRE INDUCTION REPORT" in text:
        text = "PIR"
    text = re.sub(r'^PRE\s+INDUCTION\s+REPORT', 'PIR', text)
    
    # PIR Cleanup: "PIR MO2" -> "PIR"
    text = re.sub(r'\bPIR\s+MO\d+\b', 'PIR', text)

    # MAG SEAL
    text = re.sub(r'\bREPLACEMENT\s+MAGNETIC\s+SEAL\b', 'MAG SEAL REPLACEMENT', text)
    if text == "MAG SEAL":
        text = "MAG SEAL REPLACEMENT"

    # Mag Seal Cleanup: "MO1 MAG SEAL REPLACEMENT" -> "MAG SEAL REPLACEMENT"
    text = re.sub(r'\bMO\d+\s+(MAG\s+SEAL\s+REPLACEMENT)\b', r'\1', text)

    # Bulkhead
    text = re.sub(r'^REPLACEMENT\s+OF\s+BULKHEAD\s+ASSY\s+ESN\s+(\d+)$', r'BULKHEAD ASSY \1 REPLACEMENT', text)

    # LONG TEXT MAPPING
    if "1200FH" in text and "SCHEDULE INSPECTION" in text:
        text = "SCHEDULE INSPECTION 1200FH"
    if "600FH" in text and "SCHEDULE INSPECTION" in text:
        text = "SCHEDULE INSPECTION 600FH"

    # GENERAL SN / ESN REMOVAL
    # 1. Removes "ESN 123", "SN 783", "ESN: 24432"
    text = re.sub(r'\b(ESN|SN)[:\s]*\d+\b', '', text)
    
    # CLEANUP (Parentheses)
    text = re.sub(r'\s*\([^)]*PLANNED[^)]*\)', '', text)
    
    # CONTEXTUAL LOGIC
    if text == "MO2" and "7 TU198" in orig_upper:
        text = "MO2 7 TU 198"

    is_module_only = re.match(r'^MO\d+$', text)
    if is_module_only:
        if "7 TU198" not in orig_upper:
            text = f"{text} REPLACEMENT"
    
    # Clean up extra spaces before running checks
    text = re.sub(r'\s+', ' ', text).strip()

    # MO1 VIBRATION CHECK / MO3 BORESCOPE INSPECTION -> Remove the MO prefix
    text = re.sub(r'\bMO\d+\s+(VIBRATION\s+CHECK|BORESCOPE\s+INSPECTION)\b', r'\1', text)

    # BORESCOPE Rules
    # If text is exactly "BORESCOPE" -> "BORESCOPE INSPECTION"
    if text == "BORESCOPE":
        text = "BORESCOPE INSPECTION"
    
    # MO2 PIR -> PIR 
    text = re.sub(r'\bMO\d+\s+PIR\b', 'PIR', text)

    # SUPERVISION SCHEDULED INSPECTION -> SCHEDULED INSPECTION
    text = text.replace("SUPERVISION SCHEDULED INSPECTION", "SCHEDULE INSPECTION")
    
    # POST RENTAL CHECK -> POST RENTAL INSPECTION
    text = text.replace("POST RENTAL CHECK", "POST RENTAL INSPECTION")

    # O RING REPLACEMENT MO5 -> MO5 REPLACEMENT
    if "O RING REPLACEMENT" in text and "MO5" in text:
        text = "MO5 REPLACEMENT"

    # BULKHEAD ASSY ESN REPLACEMENT -> Remove dangling "ESN"
    # This catches "ESN" if it appears as a standalone word (without digits)
    text = re.sub(r'\bESN\b', '', text)

    # Remove Serial Numbers: "ESN 2411", "S/N1234", "SERIAL NUMBER 1186", "SERIAL NUMBER: 1234"
    text = re.sub(r'\b(ESN|S/N|SERIAL\s+NUMBER)[:\s]*\d+', '', text)

    # Fix Typos
    text = text.replace("CALENDER", "CALENDAR")

    # Final trim
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove trailing periods
    text = text.rstrip('.')

    return text

# EXECUTE STANDARDIZATION
print("Standardizing Task_1...")
# Apply standardization to Task_1 using axis=1
df['Task_1'] = df.apply(standardize_task1, axis=1)

# Display Results
cols_to_show = ['Brief Description of Technical Intervention', 'Task_1', 'Task_2', 'Task_3', 'Task_4', 'Task_5', 'Task_6']
print(df[cols_to_show].head(30))
```

```python
import re
import pandas as pd

def standardize_task2_strict(text):
    """
    Standardizes Task_2 based ONLY on the specific rules provided.
    """
    if pd.isna(text):
        return text
    
    # Basic formatting: Upper case & strip whitespace
    text = str(text).upper().strip()
    
    # EXACT MATCHES / SHORT CODES
    if text == "5 REPLACEMENT": return "MO5 REPLACEMENT"
    if text == "3 REPLACEMENT": return "MO3 REPLACEMENT"
    if text == "4": return "MO4 REPLACEMENT"
    if text == "02": return "STANDARD EXCHANGE 02"
    if text == "03": return "STANDARD EXCHANGE 03"
    if text == "150H": return "150H INSPECTION"
    if text == "M05": return "MO5 INSPECTION"
    if text == "M04": return "MO4 REPLACEMENT"
    if text == "M04 REPLACEMENT": return "MO4 REPLACEMENT"
    if text == "BSI": return "BORESCOPE INSPECTION"
    if text == "PRESERVATION": return "ENGINE PRESERVATION"
    if text == "REQUIRE PRESERVATION": return "ENGINE PRESERVATION"
    if text == "VIBRATION TEST": return "VIBRATION CHECK"
    
    # SPECIFIC PHRASE MAPPINGS
    if text == "REPLACEMENT FRONT FIREWALL MODULE 01":
        return "MO1 REPLACEMENT"
    if text == "REPLACEMENT OF M01 MAGNETIC SEAL":
        return "MAG SEAL REPLACEMENT"

    # Remove leading numbering (e.g., "2. ")
    text = re.sub(r'^\d+\.\s*', '', text)
    
    # Remove "To carry out" or "Carry out" or "To" at start
    text = re.sub(r'^(TO\s+)?CARRY\s+OUT\s+', '', text)
    text = re.sub(r'^TO\s+', '', text)

    # Fix Typos
    text = text.replace("CALENDER", "CALENDAR")

    # COMPLEX LOGIC RULES

    # HMU Logic
    # Detect "Removal" AND "Installation" of HMU -> HMU REMOVAL/INSTALLATION
    if "REMOVAL" in text and "INSTALLATION" in text and "HMU" in text:
        return "HMU REMOVAL/INSTALLATION"
    
    # Detect "Installation" of HMU (but not Removal) -> HMU INSTALLATION
    if "INSTALLATION" in text and "HMU" in text and "REMOVAL" not in text:
        return "HMU INSTALLATION"

    # Engine Preservation (Catch long sentences containing 'PRESERVATIONS')
    if "ENGINE PRESERVATIONS" in text:
        return "ENGINE PRESERVATION"

    # Output Seal (Clean specific sentence)
    # "REMOVAL/INSTALLATION OF OUTPUT SEAL ESN..." -> "REMOVAL/INSTALLATION OF OUTPUT SEAL"
    if "REMOVAL/INSTALLATION OF OUTPUT SEAL" in text:
        # We strip ESNs later, so just ensure the core text is right
        pass 

    # Remove Serial Numbers: "ESN 2411", "S/N1234", "SERIAL NUMBER 1186", "SERIAL NUMBER: 1234"
    text = re.sub(r'\b(ESN|S/N|SERIAL\s+NUMBER)[:\s]*\d+', '', text)

    # Remove "AS PER TASK..." and references "(Refer : ...)"
    text = re.sub(r'\s+AS\s+PER\s+TASK.*', '', text)
    text = re.sub(r'\s*\(REFER\s*:.*\)', '', text)

    # Remove extra spaces created by deletions
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove trailing periods
    text = text.rstrip('.')
    
    return text

def split_specific_tasks(row):
    """
    Splits Task_2 into Task_2 and Task_3 ONLY for HMU and OUTPUT SEAL cases.
    """
    t2 = str(row['Task_2'])
    t3 = str(row['Task_3']) # Keep existing Task 3
    
    if pd.isna(row['Task_2']) or row['Task_2'] == 'nan':
        return t2, t3

    # CASE 1: HMU REMOVAL/INSTALLATION
    if t2 == "HMU REMOVAL/INSTALLATION":
        return "HMU REMOVAL", "HMU INSTALLATION"

    # CASE 2: REMOVAL/INSTALLATION OF OUTPUT SEAL
    # Matches "REMOVAL/INSTALLATION OF OUTPUT SEAL"
    if "REMOVAL/INSTALLATION" in t2 and "OUTPUT SEAL" in t2:
        return "OUTPUT SEAL REMOVAL", "OUTPUT SEAL INSTALLATION"

    # Default: Return original values
    return t2, t3

print("1. Standardizing Task_2...")
df['Task_2'] = df['Task_2'].apply(standardize_task2_strict)

print("2. Splitting HMU and Output Seal tasks...")
# Apply the split function row by row and update columns
df[['Task_2', 'Task_3']] = df.apply(lambda row: pd.Series(split_specific_tasks(row)), axis=1)

# Display Results
cols = ['Brief Description of Technical Intervention', 'Task_2', 'Task_3']
print("\nResults Sample:")
print(df[cols].head(20))

# Check specific examples
print("\n--- Specific Check ---")
check_vals = ["HMU REMOVAL/INSTALLATION", "OUTPUT SEAL", "HMU REMOVAL"]
print(df[df['Task_2'].str.contains('|'.join(check_vals), case=False, na=False)][cols])
```

```python
import re
import pandas as pd

def standardize_task4(text):
    """
    Standardization rules specific for Task_4 column
    """
    if pd.isna(text):
        return text
    
    # Convert to uppercase and strip whitespace
    text = str(text).upper().strip()
    
    # SB (SERVICE BULLETIN) STANDARDIZATION
    # Matches "SB" followed by digits/spaces, ignores "Ver", "Dated", "To carry out"
    # Logic: Look for "SB" + number and keep only that part
    sb_match = re.search(r'(SB\s*[\d\s]+)', text)
    if sb_match:
        # Extract "SB 123 456"
        clean_sb = sb_match.group(1).replace(" ", "") # Remove internal spaces: SB123456
        # Re-add single space for standard format: SB 123456
        clean_sb = clean_sb.replace("SB", "SB ")
        return clean_sb

    # FCU INSPECTION
    # "FCU P2 Chamber Inspection" -> "FCU INSPECTION"
    if "FCU" in text and "INSPECTION" in text:
        return "FCU INSPECTION"

    return text

# Apply standardization to Task_4
print("Standardizing Task_4...")
df['Task_4'] = df['Task_4'].apply(standardize_task4)

# Show unique values to verify
print("\nTask_4 unique values AFTER standardization:")
unique_t4 = sorted([str(t) for t in df['Task_4'].unique() if pd.notna(t) and str(t).strip() != ""])
for task in unique_t4:
    print(f"  - {task}")
```

```python
import re
import pandas as pd

def standardize_task5(text):
    """
    Standardization rules specific for Task_5 column
    """
    if pd.isna(text):
        return text
    
    # Convert to uppercase and strip whitespace
    text = str(text).upper().strip()
    
    # REMOVE REFER TEXT
    # Matches "(Refer : ...)" and removes it entirely
    # The regex looks for "(", optional "Refer", optional ":", then any text until ")"
    text = re.sub(r'\(REFER\s*:.*?\)', '', text)
    
    # If the text is now empty (e.g. the cell was ONLY the refer text), return None
    if text.strip() == "":
        return None

    # SB (SERVICE BULLETIN) STANDARDIZATION
    # Matches "SB" followed by digits and spaces.
    # We capture the numbers so we can strip the spaces inside them.
    sb_match = re.search(r'SB\s*([\d\s]+)', text)
    
    if sb_match:
        # Extract the number part: "298 73 2097"
        raw_number = sb_match.group(1)
        # Remove all spaces: "298732097"
        clean_number = raw_number.replace(" ", "")
        # Return standardized format: "SB 298732097"
        return f"SB {clean_number}"

    return text.strip()

# Apply standardization to Task_5
print("Standardizing Task_5...")
df['Task_5'] = df['Task_5'].apply(standardize_task5)

# Show unique values to verify
print("\nTask_5 unique values AFTER standardization:")
unique_t5 = sorted([str(t) for t in df['Task_5'].unique() if pd.notna(t) and str(t).strip() != ""])
for task in unique_t5:
    print(f"  - {task}")
```

```python
import re
import pandas as pd
from datetime import datetime

def parse_duration(day_str, year=None, month_str=None):
    """
    Parse various Day formats and return total duration in days.

    Handles:
    - Single days: '14', '27'
    - Simple ranges: '25 - 29', '2 to 5', '8,9', '2&3'
    - Day-Month single: '12-Nov', '3-May-20'
    - Day range with month: '10 - 11 Mar 21', '25 - 29 Apr2022'
    - Cross-month range: '12 FEB- 18 MAR', '30 MAR -3 APR', '28 - 2 OCT 20'
    - Multiple disjoint ranges: '9-Jun-23 & 7-Jul-23', '17 - 18 Jan, 05 - 07 Feb 24'

    Args:
        day_str: The Day column value
        year: Year from Year column (for context)
        month_str: Month from Mth column (for context)
    
    Returns:
        (duration, start_date, end_date)
    """
    if pd.isna(day_str):
        return None, None, None

    # Handle datetime objects
    if isinstance(day_str, datetime):
        return 1, day_str.day, day_str.day

    day_str = str(day_str).strip()
    day_str_upper = day_str.upper()

    # Month mapping
    month_map = {
        'JAN': 1, 'JANUARY': 1,
        'FEB': 2, 'FEBRUARY': 2,
        'MAR': 3, 'MARCH': 3,
        'APR': 4, 'APRIL': 4,
        'MAY': 5,
        'JUN': 6, 'JUNE': 6,
        'JUL': 7, 'JULY': 7,
        'AUG': 8, 'AUGUST': 8,
        'SEP': 9, 'SEPT': 9, 'SEPTEMBER': 9,
        'OCT': 10, 'OCTOBER': 10,
        'NOV': 11, 'NOVEMBER': 11,
        'DEC': 12, 'DECEMBER': 12
    }
    
    month_names = ['', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
                   'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

    # Get context month number
    context_month = None
    if month_str:
        month_str_upper = str(month_str).strip().upper()
        context_month = month_map.get(month_str_upper)

    # Get context year
    context_year = None
    if year:
        try:
            context_year = int(year)
            if context_year < 100:
                context_year += 2000
        except:
            context_year = 2024
    else:
        context_year = 2024

    # Check for multiple complete date segments (each with month info)
    # Pattern: "9-Jun-23 & 7-Jul-23" or "17 - 18 Jan, 05 - 07 Feb 24"
    compound_pattern = r'(\d{1,2}[^&,]+[A-Z]{3}[^&,]*)\s*[&,]\s*(\d{1,2}[^&,]+[A-Z]{3}[^&,]*)'
    if re.search(compound_pattern, day_str_upper):
        # Split and sum durations
        segments = re.split(r'\s*(?:&|AND)\s*|\s*,\s*(?=\d{1,2}\s*[-\s])', day_str_upper)
        total_duration = 0
        first_start = None
        last_end = None

        for seg in segments:
            seg = seg.strip()
            if not seg:
                continue
            dur, start, end = parse_single_segment(seg, context_year, context_month, month_map, month_names)
            if dur:
                total_duration += dur
                if first_start is None:
                    first_start = start
                last_end = end

        return total_duration if total_duration > 0 else None, first_start, last_end

    # Otherwise parse as single segment
    return parse_single_segment(day_str_upper, context_year, context_month, month_map, month_names)


def parse_single_segment(seg, year, context_month, month_map, month_names):
    """
    Parse a single date/range segment.
    Returns (duration, start_date, end_date)
    """
    seg = seg.strip()

    # Pattern 1: Single number (e.g., "14")
    if re.match(r'^\d{1,2}$', seg):
        day = int(seg)
        return 1, day, day

    # Pattern 2: Simple range "X - Y" or "X TO Y" (days only, no month)
    match = re.match(r'^(\d{1,2})\s*(?:-|TO|T0|,|&|\s)\s*(\d{1,2})$', seg)
    if match:
        start, end = int(match.group(1)), int(match.group(2))
        if end >= start:
            return end - start + 1, start, end
        else:
            # Month wrap (e.g., "29 TO 3" means crossing month boundary)
            return (30 - start) + end + 1, start, end

    # Pattern 3: Full date single (e.g., "3-MAY-20", "29-JUL-20", "9-FEB-24")
    match = re.match(r'^(\d{1,2})-(\w{3})-(\d{2,4})$', seg)
    if match:
        return 1, int(match.group(1)), int(match.group(1))

    # Pattern 4: Day-Month single (e.g., "12-NOV", "13 DEC", "12 NOV")
    match = re.match(r'^(\d{1,2})\s*-?\s*([A-Z]{3})$', seg)
    if match:
        return 1, int(match.group(1)), int(match.group(1))

    # Pattern 5: Day range with month at end (e.g., "10 - 11 MAR 21", "25 - 29 APR2022", "28 - 2 OCT 20")
    # If start_day > end_day, start is from previous month
    match = re.match(r'^(\d{1,2})\s*-\s*(\d{1,2})\s*([A-Z]{3})\s*(\d{2,4})?$', seg)
    if match:
        start_day, end_day = int(match.group(1)), int(match.group(2))
        end_month_str = match.group(3)
        yr_str = match.group(4)
        
        yr = int(yr_str) if yr_str else year
        if yr and yr < 100:
            yr += 2000
        if not yr:
            yr = 2024
            
        end_month = month_map.get(end_month_str)
        
        if start_day > end_day and end_month:
            # Start day is from previous month (e.g., "28 - 2 OCT" = 28 SEP - 2 OCT)
            start_month = end_month - 1 if end_month > 1 else 12
            start_year = yr if end_month > 1 else yr - 1
            start_month_str = month_names[start_month]
            
            try:
                start_date = datetime(start_year, start_month, start_day)
                end_date = datetime(yr, end_month, end_day)
                duration = (end_date - start_date).days + 1
                return duration, f"{start_day} {start_month_str}", f"{end_day} {end_month_str}"
            except:
                pass
        
        # Normal case: same month
        return end_day - start_day + 1, start_day, end_day

    # Pattern 6: Cross-month range with both months (e.g., "12 FEB- 18 MAR", "30 MAR -3 APR")
    match = re.match(
        r'^(\d{1,2})\s*([A-Z]{3})?\s*-\s*(\d{1,2})\s*([A-Z]{3})\s*(\d{2,4})?$',
        seg
    )
    if match:
        start_day = int(match.group(1))
        start_month_str = match.group(2)
        end_day = int(match.group(3))
        end_month_str = match.group(4)
        yr_str = match.group(5)

        yr = int(yr_str) if yr_str else year
        if yr and yr < 100:
            yr += 2000
        if not yr:
            yr = 2024

        # Get month numbers
        end_month = month_map.get(end_month_str)
        
        if start_month_str:
            start_month = month_map.get(start_month_str)
        elif start_day > end_day and end_month:
            # Infer previous month if start_day > end_day
            start_month = end_month - 1 if end_month > 1 else 12
        else:
            start_month = end_month

        if start_month and end_month:
            try:
                start_year = yr
                end_year = yr
                
                # Handle year transition (e.g., DEC to JAN)
                if start_month > end_month:
                    end_year = yr + 1
                # Handle case where start is previous year (e.g., 28 DEC - 2 JAN with end month JAN)
                if not start_month_str and start_day > end_day and end_month == 1:
                    start_month = 12
                    start_year = yr - 1
                
                start_date = datetime(start_year, start_month, start_day)
                end_date = datetime(end_year, end_month, end_day)
                duration = (end_date - start_date).days + 1
                
                start_mon_name = start_month_str if start_month_str else month_names[start_month]
                return duration, f"{start_day} {start_mon_name}", f"{end_day} {end_month_str}"
            except:
                pass

        return end_day - start_day + 1 if end_day >= start_day else None, start_day, end_day

    # Pattern 7: Long ranges with both months (e.g., "29 MAR -8 JUN 23", "30 MAR - 7 JUN 23")
    match = re.match(
        r'^(\d{1,2})\s*([A-Z]{3})\s*-\s*(\d{1,2})\s*([A-Z]{3})\s*(\d{2,4})?$',
        seg
    )
    if match:
        start_day, start_mon = int(match.group(1)), match.group(2)
        end_day, end_mon = int(match.group(3)), match.group(4)
        yr_str = match.group(5)

        yr = int(yr_str) if yr_str else year
        if yr and yr < 100:
            yr += 2000
        if not yr:
            yr = 2024

        try:
            start_date = datetime(yr, month_map[start_mon], start_day)
            end_date = datetime(yr, month_map[end_mon], end_day)
            if end_date < start_date:
                end_date = datetime(yr + 1, month_map[end_mon], end_day)
            duration = (end_date - start_date).days + 1
            return duration, f"{start_day} {start_mon}", f"{end_day} {end_mon}"
        except:
            return None, None, None

    # Pattern 8: Enumerated days with month (e.g., "10, 12, 13, 16 & 17 DEC 24")
    match = re.match(r'^([\d,\s&]+)\s*([A-Z]{3})\s*(\d{2,4})?$', seg)
    if match:
        days_str = match.group(1)
        days = [int(d.strip()) for d in re.split(r'[,&\s]+', days_str) if d.strip().isdigit()]
        if days:
            return len(days), min(days), max(days)

    # Pattern 9: Day with spaces (e.g., "5 12" meaning days 5 to 12)
    match = re.match(r'^(\d{1,2})\s+(\d{1,2})$', seg)
    if match:
        start, end = int(match.group(1)), int(match.group(2))
        if end >= start:
            return end - start + 1, start, end

    # Fallback: try to extract first and last numbers
    days = re.findall(r'\b(\d{1,2})\b', seg)
    if len(days) >= 2:
        start, end = int(days[0]), int(days[-1])
        if end >= start:
            return end - start + 1, start, end
        else:
            return (30 - start) + end + 1, start, end
    elif len(days) == 1:
        return 1, int(days[0]), int(days[0])

    return None, None, None

print("Duration parsing functions defined!")
```

```python
# Apply the parsing function to create duration, start_date, end_date
def apply_parse_duration(row):
    duration, start, end = parse_duration(row['Day'], row.get('Year'), row.get('Mth'))
    return pd.Series({'duration': duration, 'start_date': start, 'end_date': end})

# Apply to dataframe
result = df_mod1.apply(apply_parse_duration, axis=1)
df_mod1['duration'] = result['duration']
df_mod1['start_date'] = result['start_date']
df_mod1['end_date'] = result['end_date']

print("Duration parsing complete!")
print(f"\nTotal rows: {len(df_mod1)}")
print(f"Rows with duration: {df_mod1['duration'].notna().sum()}")
print(f"Rows without duration: {df_mod1['duration'].isna().sum()}")

print("\nSample results:")
print(df_mod1[['Day', 'start_date', 'end_date', 'duration']].head(50))
```

```python
# Calculate task count and task duration
import math

def count_tasks(row):
    """Count non-null tasks in Task_1 through Task_6"""
    task_cols = ['Task_1', 'Task_2', 'Task_3', 'Task_4', 'Task_5', 'Task_6']
    return sum(1 for col in task_cols if pd.notna(row[col]) and str(row[col]).strip().lower() != 'none') 

def calculate_task_duration(row):
    """
    Calculate duration per task by dividing total duration by task count.
    Round up with minimum of 1 day.
    """
    if pd.isna(row['duration']) or row['duration'] == 0:
        return None
    
    task_count = row['task_count']
    if task_count == 0:
        return None
    
    # Divide duration by task count and round up
    task_duration = math.ceil(row['duration'] / task_count)
    
    # Ensure minimum of 1 day
    return max(1, task_duration)

# Add task_count column
df_mod2['task_count'] = df_mod2.apply(count_tasks, axis=1)

# Add task_duration column
df_mod2['task_duration'] = df_mod2.apply(calculate_task_duration, axis=1)

print("Task count and task duration calculated!")
print("\nSample calculations:")
print(df_mod2[['Day', 'duration', 'Task_1', 'Task_2', 'Task_3', 'Task_4', 'Task_5', 'Task_6', 'task_count', 'task_duration']].head(30))

task_col_index = df_mod2.columns.get_loc('Task_6')

for col in ['task_count', 'task_duration']:
    if col in cols:
        cols.remove(col)

cols.insert(task_col_index + 1, 'task_count')
cols.insert(task_col_index + 2, 'task_duration')

# Reorder dataframe
df_mod2 = df_mod2[cols]

print("df_mod2 created with task_count and task_duration columns after Task_6")
print(df_mod2.columns.tolist())
```

```python
# Standardize Intervention level to L1, L2, L3 format
def standardize_intervention_level(level):
    """Standardize intervention level to L1, L2, or L3 format"""
    if pd.isna(level):
        return level
    
    level_str = str(level).strip().upper()
    
    # Extract the number from various formats
    # Match patterns like: L1, L2, L3, Level 1, Level 2, etc.
    match = re.search(r'[L]?(\d)', level_str)
    if match:
        number = match.group(1)
        return f'L{number}'
    
    return level

df_mod2['Intervention level'] = df_mod2['Intervention level'].apply(standardize_intervention_level)

print("Intervention level standardized!")
print("\nUnique intervention levels:")
print(df_mod2['Intervention level'].unique())
print("\nIntervention level distribution:")
print(df_mod2['Intervention level'].value_counts())
```

```python
# Standardize Engine names (keep engine family + number: ARRIEL 1, ARRIEL 2, etc.)
def standardize_engine(row):
    """Standardize engine names to: ARRIEL 1, ARRIEL 2, ARRIUS 1, ARRIUS 2, MAKILA 1, MAKILA 2, E-APU, ACCESSORIES.
    If no number is present in the engine name, infer from the Variant column (first digit)."""
    engine = row['Engine']
    variant = row['Variant']
    
    if pd.isna(engine):
        return engine
    
    engine_str = str(engine).strip().upper()
    engine_str = re.sub(r'\s+', ' ', engine_str)
    
    # Extract number from engine string (e.g., "ARRIEL 2" -> 2, "ARRIEL2" -> 2)
    num_match = re.search(r'(\d)', engine_str)
    
    # If no number in engine name, infer from Variant column (first character)
    if num_match:
        num = num_match.group(1)
    elif pd.notna(variant):
        variant_str = str(variant).strip()
        var_match = re.search(r'^(\d)', variant_str)
        num = var_match.group(1) if var_match else None
    else:
        num = None
    
    # ARRIEL variations (including typos like ARRRIEL, ARRIEl, ARRIEL2, etc.)
    if re.search(r'AR+I+E*L', engine_str):
        return f'ARRIEL {num}' if num else 'ARRIEL'
    
    # ARRIUS variations
    if 'ARRIUS' in engine_str or 'ARIUS' in engine_str:
        return f'ARRIUS {num}' if num else 'ARRIUS'
    
    # MAKILA variations
    if 'MAKILA' in engine_str:
        return f'MAKILA {num}' if num else 'MAKILA'
    
    # E-APU / eAPU variations
    if 'APU' in engine_str or 'E-APU' in engine_str or 'EAPU' in engine_str:
        return 'E-APU'
    
    # ACCESSORIES
    if 'ACCESSORIES' in engine_str or 'ACCESSORY' in engine_str:
        return 'ACCESSORIES'
    
    return engine_str

df_mod2['Engine'] = df_mod2.apply(standardize_engine, axis=1)

print("Engine names standardized!")
print("\nUnique Engine values:")
print(sorted(df_mod2['Engine'].unique()))
print("\nEngine distribution:")
print(df_mod2['Engine'].value_counts())
```

```python
# Standardize By column
def standardize_by(by_value):
    """Standardize By column - uppercase and extract first name"""
    if pd.isna(by_value):
        return by_value
    
    # Convert to uppercase and trim
    by_str = str(by_value).strip().upper()
    
    # For entries with "AND", keep as is (multi-person entries)
    if 'AND' in by_str:
        return by_str
    
    # Extract first word (first name or department code)
    # This handles: "JEROME METCHEDE" â†’ "JEROME", "ALI" â†’ "ALI", "BS" â†’ "BS"
    # Also handles: "RAY (TAP)" â†’ "RAY"
    first_word = by_str.split()[0]
    
    # Name corrections
    if first_word == 'AZAHARI':
        first_word = 'AZHARI'
    
    return first_word

df_mod2['By'] = df_mod2['By'].apply(standardize_by)
df_mod2['By'].unique()
```
